As seen in the value counts of isFraud column, This data is biased and I am going to do undersampling on this

```python
df['old_bal_diff_org']=df['oldbalanceOrg']-df['newbalanceOrig']
df['old_bal_diff_dest']=df['oldbalanceDest']-df['newbalanceDest']
```

```python
df_ = pd.get_dummies(df['type'],drop_first=True)
```

```python
df = pd.concat([df,df_],axis=1)
```

# Undersampling is done here to overcome the problem of biased data

```python
class_count_0, class_count_1 = y.value_counts()
class_0 = df[df['isFraud'] == 0]
class_1 = df[df['isFraud'] == 1]
print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
```

```python
class_0_under = class_0.sample(class_count_1*5)
test_under = pd.concat([class_0_under, class_1], axis=0)
test_under['isFraud'].value_counts().plot(kind='bar', title='count (target)')
```

```python
test_under.shape
```

```python
test_under.drop(['type'],axis=1)
```

```python
X = df.iloc[:test_under.shape[0]]
```

```python
X.head()
```

```python
test_under_x = X.drop(['type','nameOrig','nameDest'],axis=1)
test_under_y= test_under['isFraud']
```

```python
test_under_x.head()
```

```python
test_under_y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(test_under_x, test_under_y, test_size=0.25, random_state=42)
```

```python
ss= StandardScaler()
X_train_std = ss.fit_transform(X_train)
X_test_std = ss.transform(X_test)
```

# Hyper Parameter tuning is done here manually

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=11, random_state=42, n_jobs=2)
rf.fit(X_train_std,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = rf.predict(X_test_std)
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

As the model is giving a max of 86.96 % accuracy after training the parameters
